```python
report = df2.groupby(['home_ownership', 'term'])['loan_amnt'].agg(['sum', 'count', 'mean']).reset_index()
```

```python
report.rename(columns = {'sum of loan_amnt': 'sum',
                        'count of loans' : 'count'
                        },
              inplace = True     
             )
```

```python
report['sum of loan amount'] = report['sum']
report['count of loans'] = report['count']
report['Average of loan amount'] = report['mean']

keep_cols = ['home_ownership', 'term', 'sum of loan amount', 
             'count of loans', 'Average of loan amount']

report = report[keep_cols]
```
